```python
# Trying to find any upward trajectories
from sklearn.linear_model import LinearRegression
```

```python
# Group by car model and year to get the count of each model produced each year
model_year_counts = df.groupby(['car_model', 'year']).size().reset_index(name='count')

# Function to calculate trend for each car model
def calculate_trend(group):
    X = group['year'].values.reshape(-1, 1)
    y = group['count'].values
    model = LinearRegression().fit(X, y)
    return model.coef_[0]  # return the slope

# Apply the function to each car model and exclude grouping columns in the function
trends = model_year_counts.groupby('car_model').apply(lambda x: calculate_trend(x)).reset_index()

# Rename the columns appropriately
trends.columns = ['car_model', 'trend']

# Filter models with a positive trend
upward_trajectory_models = trends[trends['trend'] > 0]

# Sort the models by trend in descending order and get the top 5
top_5_upward_trend_models = upward_trajectory_models.sort_values(by='trend', ascending=False).head(5)

# Merge with the original DataFrame to get the car brand names
top_5_with_brands = top_5_upward_trend_models.merge(df[['car_model', 'car_brand']].drop_duplicates(), on='car_model', how='left')

# Display the top 5 models with the most significant upward trend and their brands
print(top_5_with_brands)

# Optionally, plot the trend for these top 5 models to visualize the upward trajectory
for model in top_5_with_brands['car_model']:
    model_data = model_year_counts[model_year_counts['car_model'] == model]
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=model_data['year'], y=model_data['count'], marker='o')
    plt.xlabel('Year')
    plt.ylabel('Number of Cars Manufactured')
    plt.title(f'Number of {model} Models Produced Each Year')
    plt.grid(True)
    plt.show()
```
