```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cleaned_sbt_japan.csv')
df.head(5)
```

```python
#What are the top 10 manufacturers with the most cars.
unique_manufacturers = df['car_brand'].unique()
print(unique_manufacturers)
```

```python
#gives us the number of cars each brand has manufactured
count_manufacturers = df['car_brand'].value_counts()
print(count_manufacturers)
```

```python
#gives us the max value within the car brand column
max_manufacturers = df['car_brand'].value_counts().max()
print(max_manufacturers)
```

```python
#gives us the top 10 manufacturers.
top_manufacturers = df['car_brand'].value_counts().nlargest(10)
print(top_manufacturers)
```

```python
#begin of visualization using a std fig size of (10,6)
plt.figure(figsize=(10,6))
sns.barplot(x=top_manufacturers.index, y=top_manufacturers.values, palette='viridis')
plt.title('TOP 10 CAR MANUFACTURERS IN JAPAN')
plt.xlabel('Car brands')
plt.ylabel('No. of cars manufactured')
plt.show
```

```python
#graph showing top 5 toyota models
top_toyota_models=df[df["car_brand"]== "TOYOTA"]["car_model"].value_counts().nlargest(5)
print(top_toyota_models)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_toyota_models.index, y=top_toyota_models.values, palette='mako')
plt.title('Top 5 TOYOTA Models')
plt.xlabel('Car Model')
plt.ylabel('No. of cars manufactured')
plt.show
```

```python
#No of cars by year
plt.figure(figsize=(10,6))
sns.countplot(x="year", data=df, palette="rocket")
plt.title("CAR PRODUCTION IN RECENT YEARS")
plt.xlabel("Years")
plt.ylabel("No. of cars manufactured")
plt.show
```

```python
#mileage effect on vehicles
list(df.columns)
```

```python
mileage=df['mileage_km']
main_mileage_by_year=df.groupby('year')['mileage_km'].mean()
print(main_mileage_by_year)
```

```python
plt.figure(figsize=(10,6))
plt.plot(main_mileage_by_year.index,main_mileage_by_year.values,marker='o',linestyle='-')
plt.title('MILEAGE PER YEAR')
plt.xlabel('Year')
plt.ylabel('Mileage per KM')
plt.show
```

```python
#what is the most common fuel type?
fuel=df['fuel_type'].value_counts()
fuel
```

```python
plt.figure(figsize=(10,6))
plt.title('AVAILABLE FUEL TYPES')
plt.xlabel('fuel type')
plt.ylabel('No. of cars')
sns.barplot(x=fuel.index,y=fuel.values,palette='rocket')
plt.show
```

```python
drive=df['drive_train'].value_counts(normalize=True)*100
drive
```

```python
plt.figure(figsize=(10,6))
plt.title('WHEEL DRIVE COMPARISON')
plt.xlabel('drive type')
plt.ylabel('% of cars')
sns.barplot(x=drive.index,y=drive.values,palette='mako')
```

```python
plt.figure(figsize=(6,6))
plt.title('AVAILABLE FUEL TYPES')
colours=sns.color_palette('pastel')[0:5]
plt.pie(drive,labels=drive.index,colors=colours)
plt.show()
```

```python
#Calculating mean price per engine size
mn=df.groupby('engine_size_cc')['price(Ksh)'].mean()
mn
```

```python
plt.figure(figsize=(15,12))
plt.plot(mn.index,mn.values,marker='o',linestyle='-')
plt.xticks(rotation=90)
plt.title('ENGINE SIZE WITH CORRESPONDING PRICES')
plt.xlabel('Engine size-cc')
plt.ylabel('Price')
plt.show
```

```python
list(df.columns)
```

```python
#What's the average price for each manufacture
meanprice=df.groupby('car_brand')['price(Ksh)'].mean()
meanprice.head(5)
```

```python
plt.figure(figsize=(10,6))
plt.title('AVERAGE PRICE FOR EACH MANUFACTURE')
plt.xlabel('Car brand')
plt.ylabel('Mean Price in Ksh (e^-6)')
sns.barplot(x=meanprice.index,y=meanprice.values,palette='rocket')
plt.xticks(rotation=90)
plt.show
```

```python
#does transmission affect car price
data=df.groupby('transmission')['price(Ksh)'].mean()
data
```

```python
plt.figure(figsize=(6,6))
plt.title('TRANSMISSION')
colours=sns.color_palette('pastel')[0:5]
plt.pie(data,labels=data.index,colors=colours)
plt.show()
```

```python
#effect of mileage on price
plt.figure(figsize=(10,6))
plt.plot(df['mileage_km'].round(),df['price(Ksh)'],marker='o',linestyle='-')
plt.title('MILEAGE VS PRICE')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.xlim(0,175000)
plt.ylim(0,13000000)
plt.show
```

```python
#correlation matrix
matrix=df[['mileage_km','price(Ksh)']].corr()
matrix
```
